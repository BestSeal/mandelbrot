```python
import sys, time
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
stdout = sys.stdout

BAILOUT = 16
totalTime = []


class Iterator:
    def __init__(self, iterations):
        for y in range(-39, 39):
            temp = []
            for x in range(-39, 39):
                t = time.time()
                i = self.mandelbrot(x/40.0, y/40.0, iterations)
                temp.append(time.time() - t)
            dotsT.append(temp)
    
    def mandelbrot(self, x, y, iterations):
        cr = y - 0.5
        ci = x
        zi = 0.0
        zr = 0.0
        i = 0

        while True:
            i += 1
            temp = zr * zi
            zr2 = zr * zr
            zi2 = zi * zi
            zr = zr2 - zi2 + cr
            zi = temp + temp + ci
          
            if zi2 + zr2 > BAILOUT:
                return i
            if i > iterations:
                return 0

            
plt.figure(figsize=(5, 5))   
iterationsList = [100, 1000, 10000, 100000, 1000000]
for iterations in iterationsList:
    print("For ", iterations, " iterations")
    dotsT = []
    t = time.time()
    Iterator(iterations)
    totalTime.append(time.time() - t)
    dotsT = np.array(dotsT)
    ax = sns.heatmap(dotsT)
    plt.show()
```
